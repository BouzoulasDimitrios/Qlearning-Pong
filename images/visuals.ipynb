```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 8), dpi=800)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

# Adjust the x-axis values for each subplot to create separation between bars
x_values = [1, 2]  # Adjust the values as needed

paddle_too_low  = [250, 400]
paddle_correct  = [300, 300]
paddle_too_high = [400, 250]

colors = ['blue', 'green']

ax1.bar(x_values, paddle_too_low , color=colors)
ax2.bar(x_values, paddle_correct , color=colors)
ax3.bar(x_values, paddle_too_high, color=colors)

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.4)  # Increase or decrease the value as needed

# Set the maximum Y-value
max_y = 512

# Set the y-axis limits for each subplot
ax1.set_ylim([0, max_y])
ax2.set_ylim([0, max_y])
ax3.set_ylim([0, max_y])

bar1 = ax1.bar(x_values, [paddle_too_low[0], 0], color=colors[0])
bar2 = ax1.bar(x_values, [0, paddle_too_low[1]], color=colors[1])

ax1.legend([bar1, bar2], ['paddle', 'ball'], loc='upper left')
ax2.legend([bar1, bar2], ['paddle', 'ball'], loc='upper left')
ax3.legend([bar1, bar2], ['paddle', 'ball'], loc='upper left')

ax1.set_ylabel('Y value')

ax1.set_title('Paddle below ball')
ax2.set_title('Same height')
ax3.set_title('Paddle above ball')

plt.show()
```
